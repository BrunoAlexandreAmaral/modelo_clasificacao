```python
import pandas as pd
import numpy as np
from google.colab import files
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score,balanced_accuracy_score,precision_score,recall_score,f1_score,roc_auc_score
```

```python
df = pd.read_excel("Bank_Personal_Loan_Modelling.xlsx")
```

```python
df = df.drop("ID",axis=1)
```

```python
df.columns
```

```python
x= df.drop("Personal_Loan", axis=1)
y = df["Personal_Loan"]
```

```python
x
```

```python
y
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2, random_state = 42)
```

```python
lr = LogisticRegression()
```

```python
lr_treinado = lr.fit(x_train,y_train)
```

```python
predito = lr_treinado.predict(x_test)
```

```python
x_test.assign(y_test = y_test).assign(predito = predito).head(30)
```

```python
print(f"Acurario do treino: {accuracy_score(y_train,lr_treinado.predict(x_train))}")
print(f"Acurario do teste: {accuracy_score(y_test,lr_treinado.predict(x_test))}")
print("="*40)
print(f"Acuracia Balanceada treino: {balanced_accuracy_score(y_train,lr_treinado.predict(x_train))}")
print(f"Acuracia Balanceada test: {balanced_accuracy_score(y_test,lr_treinado.predict(x_test))}")
print("="*40)
print(f"Recall treino: {recall_score(y_train,lr_treinado.predict(x_train))}")
print(f"Recall teste: {recall_score(y_test,lr_treinado.predict(x_test))}")
print("="*40)
print(f"PrecisÃ£o treino: {precision_score(y_train,lr_treinado.predict(x_train))}")
print(f"PrecisÃ£o teste: {precision_score(y_test,lr_treinado.predict(x_test))}")
print("="*40)
print(f"Roc AUC Treino : {roc_auc_score(y_train,lr_treinado.predict_proba(x_train)[:,1])}")
print(f"Roc AUC Test : {roc_auc_score(y_test,lr_treinado.predict_proba(x_test)[:,1])}")
```
