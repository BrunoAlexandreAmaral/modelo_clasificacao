```python
label = LabelEncoder()
```

```python
df.test = label.fit_transform(df.test)
```

```python
x = df[['media_rating', 'media_pedidos', 'test']]
y = df['cliente_comprou']
```

```python
model = tree.DecisionTreeClassifier()
```

```python
model_treinado = model.fit(x,y)
```

```python
fig = plt.figure(figsize=(10,10))
fig = tree.plot_tree (model_treinado, feature_names = x.columns, class_names=["Yes","No"],filled=True)
```

```python
x = x.drop(columns=["valor_real"])
```

```python
model_predito = model.predict(x)
```

```python
x["valor_predito"] = model_predito
```

```python
x["valor_real"] = y
```
