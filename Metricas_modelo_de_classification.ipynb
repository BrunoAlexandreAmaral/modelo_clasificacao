```python
import pandas as pd
import numpy as mp
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from google.colab import files
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score,balanced_accuracy_score,precision_score,recall_score,f1_score,roc_auc_score
```

```python
uploaded = files.upload()
```

```python
df = pd.read_excel("churn_data.xlsx")
```

```python
df.head()
```

```python
df.info()
```

```python
df["TotalCharges"] = pd.to_numeric(df["TotalCharges"], errors ="coerce")
```

```python
df.dropna(inplace=True)
```

```python
df.columns
```

```python
le = LabelEncoder()
```

```python
le.fit(df.Churn)
```

```python
df.Churn = le.fit_transform(df.Churn)
```

```python
colunas_caracs = df[['PhoneService', 'Contract', 'PaperlessBilling','PaymentMethod']]
colunas_numericas = df[["tenure","MonthlyCharges","TotalCharges"]]
```

```python
colunas_caracs = pd.get_dummies(colunas_caracs)
```

```python
colunas_caracs = colunas_caracs.astype({"PhoneService_No":"int","PhoneService_Yes":"int","Contract_Month-to-month":"int","Contract_One year":"int","Contract_Two year":"int","PaperlessBilling_No":"int","PaperlessBilling_Yes":"int","PaymentMethod_Bank transfer":"int","PaymentMethod_Credit card":"int","PaymentMethod_Electronic check":"int","PaymentMethod_Mailed check":"int"})
```

```python
colunas_caracs
```

```python
colunas_numericas.info()
```

```python
x = pd.concat([colunas_caracs, colunas_numericas], axis=1)
y = df["Churn"]
```

```python
x
```

```python
x.dropna(inplace=True)
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2)
```

```python
model = LogisticRegression()
```

```python
model_treinado = model.fit(x_train,y_train)
```

```python
pred_x_test = model_treinado.predict(x_test)
```

```python
x_test.assign(y_test = y_test).assign(valor_predito = pred_x_test)
```

```python
print(f"Acuracia treino:{accuracy_score(y_train, model_treinado.predict(x_train))}")
print(f"acuracia teste:{accuracy_score(y_test,model_treinado.predict(x_test))}")
print("="*40)

print(f"Acuracua Balanceada treino:{balanced_accuracy_score(y_train, model_treinado.predict(x_train))}")
print(f"Acuracia Balanceada teste:{balanced_accuracy_score(y_test,model_treinado.predict(x_test))}")
print("="*40)

print(f"Recall treino:{recall_score(y_train,model_treinado.predict(x_train))}")
print(f"Recall teste:{recall_score(y_test,model_treinado.predict(x_test))}")
print("="*40)

print(f"Precisao treino:{precision_score(y_train,model_treinado.predict(x_train))}")
print(f"Precisao de teste:{precision_score(y_test,model_treinado.predict(x_test))}")
print("="*40)

print(f"F1 score treino:{f1_score(y_train,model_treinado.predict(x_train))}")
print(f"F1 score teste:{f1_score(y_test,model_treinado.predict(x_test))}")
print("="*40)

print(f"Rocauc Treino:{roc_auc_score(y_train,model_treinado.predict_proba(x_train)[:,1])}")
print(f"Rocauc test:{roc_auc_score(y_test,model_treinado.predict_proba(x_test)[:,1])}")
```
