```python
import pandas as pd
import numpy as np
from google.colab import files
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
```

```python
df = pd.read_excel("golf_df.xlsx")
```

```python
df.head()
```

```python
le = LabelEncoder()
```

```python
le.fit(df.clima)
df.clima = le.transform(df.clima)

le.fit(df.temp)
df.temp = le.transform(df.temp)

le.fit(df.umidade)
df.umidade = le.transform(df.umidade)

le.fit(df.vento)
df.vento = le.transform(df.vento)

le.fit(df.jogar)
df.jogar = le.transform(df.jogar)
```

```python
df
```

```python
x =df.iloc[:,0:-1]
y = df.iloc[:,-1]
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2)
```

```python
model =LogisticRegression()
```

```python
model_treinado = model.fit(x_train,y_train)
```

```python
predito = model.predict(x_test)
```

```python
x_test["real"] = y_test
```
