```python
model_label = LabelEncoder()
```

```python
df.clima = model_label.fit_transform(df.clima)
df.temp = model_label.fit_transform(df.temp)
df.umidade = model_label.fit_transform(df.umidade)
df.vento = model_label.fit_transform(df.vento)
df.jogar = model_label.fit_transform(df.jogar)
```

```python
df
```

```python
x = df[["clima","temp","umidade","vento"]]
y = df["jogar"]
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y, test_size=0.2)
```

```python
model_lr = LogisticRegression()
```

```python
model_lr_treinado = model_lr.fit(x_train,y_train)
```

```python
predicao_train = model_lr.predict(x_train)
```

```python
x_train["valor_real"] = y_train
```

```python
x_train["valor_predito"] = predicao_train
```
